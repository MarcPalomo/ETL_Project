```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Medical_Doctors.csv"
doctors_df = pd.read_csv(csv_file)
doctors_df.head()
```

```python
doctors_df2 = doctors_df.loc[doctors_df["Year"]>=2000]
doctors_df2.head()
```

```python
len(doctors_df2["Country"].unique())
```

```python
# Create a filtered dataframe from specific columns
doctors_cols = ["Country", "Year", "Medical doctors (per 10 000 population)", "Medical doctors (number)"]
doctors_transformed= doctors_df2[doctors_cols].copy()

# Rename the column headers
doctors_transformed = doctors_transformed.rename(columns={"Country": "Country",
                                                          "Year": "Year",
                                                          "Medical doctors (per 10 000 population)": "DoctorsPerTenThousand", 
                                                         "Medical doctors (number)": "Doctors"})
```

```python
doctors_transformed.set_index("Country", inplace=True)
```

```python
doctors_transformed.head()
```

```python
connection_string = "root:<password>@localhost/ETLProject_db"
engine = create_engine(f'mysql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
doctors_transformed.to_sql(name='doctors', con=engine, if_exists='append', index=True)
```

```python
csv_file = "Malaria_Deaths.csv"
deaths_df = pd.read_csv(csv_file, skiprows=[0])
deaths_df.head()
```

```python
deaths_df2 = pd.melt(deaths_df, id_vars=["Country"], value_vars=["2017", "2016", "2015", "2014", 
                                                                 "2013", "2012", "2011", "2010", 
                                                                 "2009", "2008", "2007", "2006", 
                                                                 "2005", "2004", "2003", "2002", 
                                                                 "2001", "2000"])
deaths_df2.head()
```

```python
deaths_df2.sort_values(by=["Country","variable"], ascending=True, inplace=True)
deaths_df2.head()
```

```python
len(deaths_df2["Country"].unique())
```

```python
# Create a filtered dataframe from specific columns
deaths_cols = ["Country", "variable", "value"]
deaths_transformed= deaths_df2[deaths_cols].copy()

# Rename the column headers
deaths_transformed = deaths_transformed.rename(columns={"Country": "Country",
                                                          "variable": "Year",
                                                          "value": "Deaths"})
```

```python
deaths_transformed.head()
```

```python
deaths_transformed.set_index("Country", inplace=True)
deaths_transformed.head()
```

```python
deaths_transformed.to_sql(name='malaria', con=engine, if_exists='append', index=True)
```
